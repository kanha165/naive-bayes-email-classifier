```python
import pickle
import matplotlib.pyplot as plt
```

```python
model = pickle.load(open("naive_bayes_model.pkl", "rb"))
cv = pickle.load(open("vectorizer.pkl", "rb"))

email_text = input("Enter an email message to test: ")

email_vector = cv.transform([email_text])

prediction = model.predict(email_vector)
probability = model.predict_proba(email_vector)[0]   # [spam_prob, not_spam_prob]

from IPython.display import display, HTML

if prediction[0] == 1:
    display(HTML(f"<h2 style='color:red;'>⚠️ This email is SPAM</h2>"))
else:
    display(HTML(f"<h2 style='color:green;'>✅ This email is NOT SPAM</h2>"))

labels = ["Not Spam", "Spam"]
plt.bar(labels, probability, color=["green", "red"])
plt.title("Spam Detection Probability")
plt.ylabel("Confidence Level")
plt.ylim(0, 1)
for i, v in enumerate(probability):
    plt.text(i, v + 0.02, f"{v*100:.2f}%", ha='center', fontsize=12)
plt.show()

```
